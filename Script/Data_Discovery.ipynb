```python
import pandas as pd
from matplotlib import pyplot as plt
import json
```

```python
# Main CSV
```

```python
csv_path = '../Data/USvideos.csv'
df = pd.read_csv(csv_path)
df.info()
```

```python
df.head()
```

```python
# Catagory df from JSON
```

```python
json_path = '../Data/US_category_id.json'
with open(json_path) as f:
    data = json.load(f)
```

```python
title = []
ID = []

for record in data['items']:
    cata_title = record['snippet']['title']
    cata_ID = record['id']
    title.append(cata_title)
    ID.append(cata_ID)
```

```python
category_df = pd.DataFrame({'CategoryName': title, 'id': ID})
category_df['id'] = category_df['id'].astype('int64')
category_df.info()
```

```python
# Joining both DFs
```

```python
master_table = df.merge(category_df, how='inner', left_on='category_id', right_on='id')
master_table.head()
```

```python
# Data discovery
grouped_table = master_table.groupby(['CategoryName']).views.mean().reset_index()
grouped_table.round(3)
```

```python
# data discovery graphing
x = grouped_table['views']
y = grouped_table['CategoryName']

plt.bar(y, x)
plt.xticks(rotation=90)
plt.show()
```
